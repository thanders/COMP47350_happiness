```python
# Import package for reading csv files 
import pandas as pd
# Import package for numpy
import numpy as np

# Reading from a csv file, into a data frame
df = pd.read_csv('Final_2008-2017.csv')
df.drop(['GINI index (World Bank estimate), average 2000-15', 'GINI index (World Bank estimate)'],axis=1)
```

```python
df_file = {2008: ['2008.csv'],
           2009: ['2009.csv'],
           2010: ['2010.csv'],
           2011: ['2011.csv'],
           2012: ['2012.csv'],
           2013: ['2013.csv'],
           2014: ['2014.csv'],
           2015: ['2015.csv'],
           2016: ['2016.csv'],
           2017: ['2017.csv']}
for year in range(2008, 2018):
    df_file[year].append(df.loc[df['year'] == year])
    new_file = df_file[year][0]
    df_file[year][1].to_csv(new_file)
```
