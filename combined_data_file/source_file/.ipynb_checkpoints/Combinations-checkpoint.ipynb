```python
# Import package for reading csv files 
import pandas as pd
# Import package for numpy
import numpy as np

# Reading from a csv file, into a data frame
df = pd.read_csv('Dying_baby_data.csv')
df2 = pd.read_csv('WHOLifeExpectancy.csv')
df3 = pd.read_csv('WHR151617.csv')
df4 = pd.read_csv('GDP_in_2011USD.csv')
df5 = pd.read_csv('transparency_international_CPI.csv')

# The following are Crime Index For country year.csv
cdf1 = pd.read_csv('Crime Index for Country 2012.csv')
cdf2 = pd.read_csv('Crime Index for Country 2013.csv')
cdf3 = pd.read_csv('Crime Index for Country 2014.csv')
cdf4 = pd.read_csv('Crime Index for Country 2015.csv')
cdf5 = pd.read_csv('Crime Index for Country 2016.csv')
cdf6 = pd.read_csv('Crime Index for Country 2017.csv')
# The following one is happiness&descriptivefeat_2008-2017.csv. we will combine all data set into this file
dfn = pd.read_csv('h2008-2017.csv')
```

```python
# Insert CPI score into h2008-2017, df5
newRows = [0]*1562
dfn['CPI'] = newRows
year_index = {2012: 'CPI Score 2012',
              2013: 'CPI Score 2013',
              2014: 'CPI score 2014',
              2015: 'CPI score 2015',
              2016: 'CPI score 2016',
              2017: 'CPI score 2017'}

for year in [2012,2013,2014,2015,2016,2017]:
    for i in df5.index:
        country = df5['Country'].loc[i]
        #print(year, year_index[year])
        cpi = df5[str(year_index[year])].loc[i]
        for j in dfn.loc[dfn['year']==year].index:
            if dfn['country'].loc[j] == country:
                dfn['CPI'].loc[j] = cpi
```

```python
# Insert Crime Index for Country into h2008-2017.csv, cdf1-6
newRows = [0]*1562
crime1 = 'Crime Index'
crime2 = 'Safety Index'
for column in [crime1, crime2]:
    dfn[column] = newRows

crime_dict = {2012: cdf1,
              2013: cdf2,
              2014: cdf3,
              2015: cdf4,
              2016: cdf5,
              2017: cdf6}
for year in [2012,2013,2014,2015,2016]:
    for i in crime_dict[year].index:
        country = crime_dict[year]['Country'].loc[i]
        c1 = crime_dict[year][crime1].loc[i]
        c2 = crime_dict[year][crime2].loc[i]
        for j in dfn.loc[dfn['year']==year].index:
            if dfn['country'].loc[j] == country:
                dfn[crime1].loc[j] = c1
                dfn[crime2].loc[j] = c2
                
```

```python
# Insert GDP into h2008-2017.csv, df4
newRows = [0]*1562
dfn['GDP'] = newRows
for year in [2015, 2016, 2017]:
    for i in df4.index:
        country = df4['Country'].loc[i]
        gdp = df4[str(year)].loc[i]
        #print(country, gdp)
    
        for j in dfn.loc[dfn['year']==year].index:
            if dfn['country'].loc[j] == country:
                #dfn['GDP'] = gdp
                #print(country, gdp, year)
                dfn['GDP'].loc[j] = gdp
```

```python
# insert WHR151617 into h2008-2017.csv, only 2015,2016 and 2017, df3
newRows = [0]*1562
WHR1 = 'Happiness Rank'
WHR2 = 'Happiness Score'
WHR3 = 'Standard Error'
WHR4 = "Lower Confidence Interval"
WHR5 = 'Upper Confidence Interval'
WHR6 = 'Economy (GDP per Capita)'
WHR7 = 'Family'
WHR8 = 'Health (Life Expectancy)'
WHR9 = 'Freedom'
WHR10 = 'Trust (Government Corruption)'
WHR11 = 'Generosity'
WHR12 = 'Dystopia Residual'
for column in [WHR1, WHR2, WHR3, WHR4, WHR5, WHR6, WHR7, WHR8, WHR9, WHR10, WHR11, WHR12]:
    dfn[column] = newRows
    
for year in [2015,2016,2017]:
    for i in df3.index:
        if str(year) in str(df3['Year'].loc[i]):
            country = df3['Country'].loc[i]
            c1 = df3[WHR1].loc[i]
            c2 = df3[WHR2].loc[i]
            c3 = df3[WHR3].loc[i]
            c4 = df3[WHR4].loc[i]
            c5 = df3[WHR5].loc[i]
            c6 = df3[WHR6].loc[i]
            c7 = df3[WHR7].loc[i]
            c8 = df3[WHR8].loc[i]
            c9 = df3[WHR9].loc[i]
            c10 = df3[WHR10].loc[i]
            c11 = df3[WHR11].loc[i]
            c12 = df3[WHR12].loc[i]
            for j in dfn.loc[dfn['year']==year].index:
                if dfn['country'].loc[j] == country:
                    dfn.loc[j, [WHR1]]= c1
                    dfn.loc[j, [WHR2]]= c2
                    dfn.loc[j, [WHR3]]= c3
                    dfn.loc[j, [WHR4]]= c4
                    dfn.loc[j, [WHR5]]= c5
                    dfn.loc[j, [WHR6]]= c6
                    dfn.loc[j, [WHR7]]= c7
                    dfn.loc[j, [WHR8]]= c8
                    dfn.loc[j, [WHR9]]= c9
                    dfn.loc[j, [WHR10]]= c10
                    dfn.loc[j, [WHR11]]= c11
                    dfn.loc[j, [WHR12]]= c12
```

```python
# insert under five mortality into h2008-2017.csv, df (the last year in this table is ended on 2016)
newRows = [0]*1562
dfn['underFiveMortality'] = newRows
for year in [2008,2009,2010,2011,2012,2013,2014,2015,2016]:
    for i in df.loc[(df['Year']==year)].index:
        country = df['Country'].loc[i]
        underFiveMortality = df['Under-five mortality rate (probability of dying by age 5 per 1000 live births)'].loc[i]
    
        for j in dfn.loc[dfn['year']==year].index:
            if dfn['country'].loc[j] == country:
                dfn.loc[j, ['underFiveMortality']]= underFiveMortality
```

```python
# insert Life expectancey at birth into h2008-2017.csv, df2
column1 = 'Life expectancy at birth (years)'
column2 = 'Life expectancy at age 60 (years)'

#df2.dtypes
# init the new columns on 1562 lines
for column in [column1, column2]:
    dfn[column] = newRows
for year in [2008,2009,2010,2011,2012,2013,2014,2015]:
    for i in df2.index:
        if str(year) in df2['Unnamed: 1'].loc[i]:
            country = df2['Unnamed: 0'].loc[i]
            col1 = df2[column1].loc[i]
            col2 = df2[column2].loc[i]
            for j in dfn.loc[dfn['year']==year].index:
                if dfn['country'].loc[j] == country:
                    dfn.loc[j, [column1]]= col1
                    dfn.loc[j, [column2]]= col2
```
